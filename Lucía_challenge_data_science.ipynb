### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calculamos el ingreso total por tienda sumando la columna 'Precio'
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostramos los resultados
print("Ingreso total Tienda 1: $", ingreso_tienda1)
print("Ingreso total Tienda 2: $", ingreso_tienda2)
print("Ingreso total Tienda 3: $", ingreso_tienda3)
print("Ingreso total Tienda 4: $", ingreso_tienda4)
```

```python
import matplotlib.pyplot as plt

# Supongamos que tienes el ingreso total por tienda calculado (como en pasos anteriores)
ingreso_total_t1 = tienda['Precio'].sum()
ingreso_total_t2 = tienda2['Precio'].sum()
ingreso_total_t3 = tienda3['Precio'].sum()
ingreso_total_t4 = tienda4['Precio'].sum()

# Crear una lista con los nombres de las tiendas y los ingresos totales
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total_t1, ingreso_total_t2, ingreso_total_t3, ingreso_total_t4]

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(tiendas, ingresos, color=['lightblue', 'lightgreen', 'salmon', 'lightcoral'])

# Agregar título y etiquetas
plt.title('Ingreso Total por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Ingreso Total (en $)', fontsize=12)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Cantidad de productos vendidos por categoría en cada tienda

ventas_categoria_t1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_categoria_t2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_categoria_t3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_categoria_t4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)

# Mostrar resultados
print("Ventas por categoría - Tienda 1:\n", ventas_categoria_t1, "\n")
print("Ventas por categoría - Tienda 2:\n", ventas_categoria_t2, "\n")
print("Ventas por categoría - Tienda 3:\n", ventas_categoria_t3, "\n")
print("Ventas por categoría - Tienda 4:\n", ventas_categoria_t4, "\n")
```

```python
import matplotlib.pyplot as plt

# Contamos las ventas por categoría (como en el paso anterior)
ventas_categoria_t1 = tienda.groupby('Categoría del Producto').size()
ventas_categoria_t2 = tienda2.groupby('Categoría del Producto').size()
ventas_categoria_t3 = tienda3.groupby('Categoría del Producto').size()
ventas_categoria_t4 = tienda4.groupby('Categoría del Producto').size()

# Crear el gráfico de dispersión para Tienda 1
plt.figure(figsize=(8, 6))
plt.scatter(ventas_categoria_t1.index, ventas_categoria_t1.values, color='blue', label='Tienda 1')

# Crear el gráfico de dispersión para Tienda 2
plt.scatter(ventas_categoria_t2.index, ventas_categoria_t2.values, color='green', label='Tienda 2')

# Crear el gráfico de dispersión para Tienda 3
plt.scatter(ventas_categoria_t3.index, ventas_categoria_t3.values, color='red', label='Tienda 3')

# Crear el gráfico de dispersión para Tienda 4
plt.scatter(ventas_categoria_t4.index, ventas_categoria_t4.values, color='purple', label='Tienda 4')

# Agregar título y etiquetas
plt.title('Ventas por Categoría por Tienda', fontsize=16)
plt.xlabel('Categoría de Producto', fontsize=12)
plt.ylabel('Número de Ventas', fontsize=12)
plt.xticks(rotation=45)  # Rotar etiquetas en el eje X para que no se superpongan

# Agregar leyenda
plt.legend()

# Ajustar el diseño
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular la valoración promedio por tienda
valoracion_tienda1 = tienda['Calificación'].mean()
valoracion_tienda2 = tienda2['Calificación'].mean()
valoracion_tienda3 = tienda3['Calificación'].mean()
valoracion_tienda4 = tienda4['Calificación'].mean()

# Mostrar los resultados
print("Valoración promedio Tienda 1:", valoracion_tienda1)
print("Valoración promedio Tienda 2:", valoracion_tienda2)
print("Valoración promedio Tienda 3:", valoracion_tienda3)
print("Valoración promedio Tienda 4:", valoracion_tienda4)
```

```python
import matplotlib.pyplot as plt

# Supongamos que ya tienes calculado el valor de la calificación promedio para cada tienda:
calificacion_promedio_t1 = tienda['Calificación'].mean()
calificacion_promedio_t2 = tienda2['Calificación'].mean()
calificacion_promedio_t3 = tienda3['Calificación'].mean()
calificacion_promedio_t4 = tienda4['Calificación'].mean()

# Crear una lista con los nombres de las tiendas y las calificaciones promedio
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calificacion_promedio_t1, calificacion_promedio_t2, calificacion_promedio_t3, calificacion_promedio_t4]

# Crear el gráfico de líneas
plt.figure(figsize=(8, 6))
plt.plot(tiendas, calificaciones, marker='o', color='purple', linestyle='-', linewidth=2, markersize=8)

# Agregar título y etiquetas
plt.title('Calificación Promedio por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
